# DATA CLEANING

```python
df[df.isnull().any(axis=1)]
```

```python
""" The dataset is divided into two regions... Sidi-Bel Abbes Region and some other
    So we will make a seperate column indication that"""
```

```python
df.loc[:122 , "Region"]=0
df.loc[122: , "Region"] = 1

""" Bejala Region is assigned 0 and Sidi-Bel Abbes Region is assigned 1"""
```

```python
df.info()
```

```python
df[["Region"]] = df[["Region"]].astype(int)
```

```python
df.info()
```

```python
df[df.isnull().any(axis=1)]
```

```python
df=df.dropna().reset_index(drop=True)  # Remove nan values and reset index
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df=df.drop(122).reset_index(drop=True)  #rEMOVE 122ND ROW
```

```python
df.columns
```

```python
df.columns=df.columns.str.strip()   # Fixing white spaces and inconsistent gaps
df.columns
```

```python
# First change dtype to integers where needed

df[["day" , "month" , "year" , "Temperature" , "RH" , "Ws"]] = df[["day" , "month" , "year" , "Temperature" , "RH" , "Ws"]].astype(int)
```

```python
# Now changing dtype to float where needed

df[["Rain" , "FFMC" , "DMC" , "DC" , "ISI" , "BUI" , "FWI"]] = df[["Rain" , "FFMC" , "DMC" , "DC" , "ISI" , "BUI" , "FWI"]].astype(float)
```

```python
df.info()
```

```python
# Saving cleaned dataset

df.to_csv("Algerian_forest_fires_cleaned_dataset.csv" , index=False)
```
